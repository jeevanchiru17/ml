```python
import csv

# Initialize an empty list to store the training data
a = []

# Read the training data from the CSV file
with open('enjoysport.csv', 'r') as csvfile:
    for row in csv.reader(csvfile):
        a.append(row)

# Print the training data
print(a)
print("\nThe total number of training instances are:", len(a))

# Determine the number of attributes (excluding the class label)
num_attribute = len(a[0]) - 1

# Initialize the hypothesis with the most specific value (0 represents "no information")
print("\nThe initial hypothesis is:")
hypothesis = ['0'] * num_attribute
print(hypothesis)

# Find-S algorithm to find the maximally specific hypothesis
for i in range(len(a)):
    if a[i][num_attribute] == 'yes':  # Check if the instance is positive
        for j in range(num_attribute):
            if hypothesis[j] == '0' or hypothesis[j] == a[i][j]:
                hypothesis[j] = a[i][j]  # Update hypothesis for matching attributes
            else:
                hypothesis[j] = '?'  # Generalize hypothesis if there's a mismatch
    print("\nThe hypothesis for the training instance {} is:".format(i + 1), hypothesis)

# Print the final maximally specific hypothesis
print("\nThe Maximally specific hypothesis for the training instances is:")
print(hypothesis)
```

```python
import numpy as np
import pandas as pd

# Load data
data = pd.read_csv('enjoysport.csv')

# Separate concepts and target values
concepts = np.array(data.iloc[:, 0:-1])
print("\nInstances are:\n", concepts)
target = np.array(data.iloc[:, -1])

# Map categorical target values to numerical values
target = np.array([1 if val.strip() == 'Yes' else 0 for val in target], dtype=np.int64)
print("\nTarget Values are: ", target)

def learn(concepts, target):
    specific_h = concepts[0].copy()
    print("\nInitialization of specific_h and general_h")
    print("\nSpecific Boundary: ", specific_h)
    general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]
    print("\nGeneric Boundary: ", general_h)

    # Iterate through all instances
    for i, h in enumerate(concepts):
        print("\nInstance", i + 1, "is ", h)

        # Positive instance
        if target[i] == 1:
            print("Instance is Positive")
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'
                    general_h[x][x] = '?'
        
        # Negative instance
        if target[i] == 0:
            print("Instance is Negative")
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]
                else:
                    general_h[x][x] = '?'
        
        print("Specific Boundary after ", i + 1, "Instance is ", specific_h)
        print("Generic Boundary after ", i + 1, "Instance is ", general_h)
        print("\n")
    
    return specific_h, general_h

# Learn and get the final hypotheses
specific_h, general_h = learn(concepts, target)
print("Final Specific Hypothesis: ", specific_h)
print("Final General Hypothesis: ", general_h)
```
